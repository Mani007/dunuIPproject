```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mydf = pd.read_csv('dataset.csv')
mydf
```

```python
plt.plot(mydf['States/ Indicators'], mydf['Gross Enrollment Ratio (GER) at Secondary Level 2021-22 - All'], marker='o',markeredgecolor='b')
plt.title('State wise Gross enrollment Ratio (GER) at Secondary Level for year 2021-22')
plt.xlabel('States')
plt.ylabel('Gross Enrollment Ratio (GER) at Secondary Level')
fig = plt.gcf()  # get current figure
fig.set_size_inches(14, 6)  # set new size
fig.show()
```

```python
plt.bar(mydf['States/ Indicators'], mydf['Literacy Rate - All'], color='green')
plt.title('State wise Literacy Rate - All category ')
plt.xlabel('States')
plt.ylabel('Literacy Rate - All')
fig = plt.gcf()  # get current figure
fig.set_size_inches(14, 6)  # set new size
fig.show()
```

```python
plt.barh(mydf['States/ Indicators'], mydf['Sex-Ratio - All'], color='purple')
plt.title('State wise Sex ratio - All category ')
plt.xlabel('Sex-Ratio - All')
plt.ylabel('States')
fig = plt.gcf()  # get current figure
fig.set_size_inches(14, 6)  # set new size
fig.show()
```

```python
x_ticks = np.arange(len(mydf['Literacy Rate - All']))
mywidth=0.25
#plt.xticks(x_ticks+10/2,mydf['States/ Indicators'])

plt.bar(x_ticks-mywidth,mydf['Literacy Rate - All'],width=mywidth,label='Literacy Rate - All',color='green')
#plt.bar(x_ticks,mydf['Sex-Ratio - All'],width=mywidth,label='Sex-Ratio - All',color='green')
plt.bar(x_ticks+mywidth,mydf['Gross Enrollment Ratio (GER) at Secondary Level 2021-22 - All'],width=mywidth,label='(GER)2021-22 - All',color='purple')
plt.title('State wise Sex ratio - All category ')
plt.xlabel('States')
plt.ylabel('Values')
plt.legend()
plt.show()
```
